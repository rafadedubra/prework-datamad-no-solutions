```python
import random, os
clear = lambda: os.system("cls")

def move():
    player = input("Choose a move:\n Rock = 1\n Paper = 2\nScissor = 3\nMake A Move : ")
    try :   
        player = int(player)
        if player in [1,2,3]:
            return player
    except ValueError:
        pass
    print("Choose a move:\n Rock = 1\n Paper = 2\nScissor = 3\nMake A Move : ")


  
def combat():
    playerChoice = move()
    machine = random.randint(1, 3)
    if (playerChoice == machine):
        return 0
    elif (playerChoice == 1 and machine == 2):
        print ("Your choice is Rock and the machine is Paper\n")
        return 1
    elif (playerChoice == 1 and machine == 3):
        print ("Your choice is Rock and the machine is Scissor\n")
        return 2
    elif  (playerChoice == 2 and machine == 3):
        print ("Your choice is Paper and the machine is Scissor\n")
        return 1
    elif  (playerChoice == 2 and machine == 1):
       print ("Your choice is Paper and the machine is Rock\n")
       return 2
    elif  (playerChoice == 3 and machine == 1):
      print ("Your choice is Scissor and the machine is Rock\n")
      return 1
    elif  (playerChoice == 3 and machine == 2):
       print ("Your choice is Scissor and the machine is Paper\n")
       return 2

    
def game ():
    max_game = 0
    machine_acc = 0
    player_acc = 0  
    tie=0
    while (max_game < 5):
        result = combat()
        max_game += 1
        if (result == 2):
            print("Win round Player!\n")
            player_acc += 1
        elif(result == 1):
            print ("Win round Machine!\n")
            machine_acc += 1
        elif (result == 0):
            print ("Tie\n")
            tie += 1
            
    print ("Final Results: \nThe player has won : " + str(player_acc) + " times")
    print ("The machine has won : " + str(machine_acc) + " times")
    print ("The number of ties is: " + str(tie))
    tryAgain()

def tryAgain():    
    choice = str(input("Would you like to play again? y/n\n"))
    if choice == "y":
        clear()
        game()
        
game()


  
```
